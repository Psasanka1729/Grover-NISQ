```python
import time
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, transpile
start = time.time()
```

# 3 qubit gates

```python
qr = QuantumRegister(14, 'c') # 14 control qubits.
target_qubit = QuantumRegister(1,'t') # target qubit.
anc = QuantumRegister(5, 'ancilla') # ancilla qubits.
qc = QuantumCircuit(qr,target_qubit, anc)

for i in range(4):
    qc.mcx(qr[3*i:3*i+3], anc[i])
    
qc.mcx([qr[12],qr[13],anc[0]],anc[4])

qc.mcx(anc[1:5],target_qubit[0])

qc.mcx([qr[12],qr[13],anc[0]],anc[4])

for i in range(3,-1,-1):
    qc.mcx(qr[3*i:3*i+3], anc[i])
qc.draw('mpl') #.savefig('ClassicQ.png', dpi=600)
```

```python
circuit_decomposition= transpile(qc, basis_gates=['u','cx'], optimization_level = 3)
```

```python
#circuit_decomposition.draw('mpl').savefig('ClassicQ_decompoition.png', dpi=100)
```

```python
len(circuit_decomposition)
```

```python
print("Circuit depth =", circuit_decomposition.depth())
```

```python
print("Number of 1 and 2 qubit gates =",circuit_decomposition.count_ops())
```

# 3 qubits gate version 2

```python
qr = QuantumRegister(14, 'c') # 14 control qubits.
target_qubit = QuantumRegister(1,'t') # target qubit.
anc = QuantumRegister(5, 'ancilla') # ancilla qubits.
qc = QuantumCircuit(qr,target_qubit, anc)

# Three qubits gates.
qc.mcx(qr[0:3],anc[0])
qc.mcx(qr[3:6],anc[1])
qc.mcx(qr[6:9],anc[2])
qc.mcx(qr[9:12],anc[3])
qc.ccx(qr[12],qr[13],anc[4])

qc.mcx(anc[0:5],target_qubit[0])

# Three qubits gates.
qc.ccx(qr[12],qr[13],anc[4])
qc.mcx(qr[9:12],anc[3])
qc.mcx(qr[6:9],anc[2])
qc.mcx(qr[3:6],anc[1])
qc.mcx(qr[0:3],anc[0])

qc.draw('mpl')#.savefig('ClassicQ14_3_1.png', dpi=600)
```

```python
circuit_decomposition= transpile(qc, basis_gates=['u','cx'], optimization_level = 3)
```

```python
len(circuit_decomposition)
```

```python
print("Circuit depth =", circuit_decomposition.depth())
```
