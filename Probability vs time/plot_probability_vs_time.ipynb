```julia
using DelimitedFiles, Plots
file = raw"probability_data.txt"
M1 = readdlm(file)
p01 = M1[:,1];
pxbar1 = M1[:,2];
t1 = M1[:,3];

using Plots
gr()
plot(t1,p01, lw =4, label = "P0(t)",title = "\\epsilon = 0.008, L = 8, Noise Seed = 2022",dpi = 600)
plot!(t1, p01, seriestype = :scatter,label="")
plot!(t1,pxbar1, lw =4,label = "Pxbar(t)")
plot!(t1, pxbar1, seriestype = :scatter,label="")

N = 8
hline!([[1/(2^N)]],lc=:red,linestyle = :dash,label = "1/N",legend = :outertopright)
hline!([[(2^N-1)/(2^N)]],lc=:black,linestyle = :dash,label = "(N-1)/N")
xlabel!("Number of Grover iterations")
ylabel!("Probability")
```
