```julia
using Plots
using CSV
#using LaTeXStrings
#using PyPlot
```

```julia
using DelimitedFiles, Plots
file = raw"PlotData.txt"
M = readdlm(file)
x = M[:,1]; # index
y = M[:,3]; # entropy
y_seom = M[:,4] # SEOM
eigen = M[:,2]; # SEOM
#plot(M[:,1], M[:,2])
```

```julia
MyTitle = "\\epsilon = 0.05, Noise = 2,L = 8, CMap = Eigenvalue";
```

```julia
using Plots
#plot(x, y, title = "Two Lines", label = ["Line 1" "Line 2"], lw = 3)
gr() # Set the backend to GR
p = plot(x,y, yerr = y_seom,
    seriestype = :scatter,
    #color = "dodgerblue1",
    markerstrokecolor = "grey30",
    markerstrokewidth=0.618,
    #markerstrokealpha = 0.1,
    markersize=5,
    thickness_scaling = 1.4,
    xlims=(-5, 260), 
    #ylims=(0,1),
    title = MyTitle,
    #grid=True,
    label = "",
    dpi=600,
    zcolor = eigen,
    color = :RdBu_4, #:RdBu_10,
    #colorbar = true,
    #colorbar_title="Eigenvalue",
    right_margin = 5Plots.mm,
    left_margin = Plots.mm,
    titlefontsize=11,
    guidefontsize=11,
    tickfontsize=11,
    legendfontsize=11,
    ) 

plot!(size=(1000,600)) # sets the resolution of the image.
#plot!(x,y,lw=1, color = "red", label ="" )
#rc("text",usetex="True")
xlabel!("Index of Eigenvalues")
ylabel!("Entropy")
```
