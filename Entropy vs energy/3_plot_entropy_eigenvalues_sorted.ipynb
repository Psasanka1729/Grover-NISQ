```julia
using DelimitedFiles, Plots
file = raw"PlotData.txt"
M = readdlm(file)
y = M[:,3]; # entropy
x = M[:,2]; # energy
y_seom = M[:,4] # SEOM
#eigen = M[:,2]; # SEOM
#plot(M[:,1], M[:,2])
```

```julia
MyTitle = "\\epsilon = 0.005, Noise = 3,L = 8, The bulk";
```

```julia
using Plots
#plot(x, y, title = "Two Lines", label = ["Line 1" "Line 2"], lw = 3)
gr() # Set the backend to GR
p = plot(x,y, yerr = y_seom,
    seriestype = :scatter,
    #color = "dodgerblue1",
    markerstrokecolor = "grey30",
    markerstrokewidth=0.418,
    #markerstrokealpha = 0.1,
    markersize=4,
    thickness_scaling = 1.4,
    xlims=(-1, 1), 
    #ylims=(0,1),
    title = MyTitle,
    #grid=True,
    label = "",
    dpi=600,
    #zcolor = eigen,
    color = :RdBu_4, #:RdBu_10,
    #colorbar = true,
    #colorbar_title="Eigenvalue",
    right_margin = Plots.mm,
    left_margin = Plots.mm,
    titlefontsize=11,
    guidefontsize=11,
    tickfontsize=11,
    legendfontsize=11,
    ) 

plot!(size=(800,600)) # sets the resolution of the image.
#plot!(x,y,lw=1, color = "red", label ="" )
#rc("text",usetex="True")
xlabel!("Energy")
ylabel!("Entropy")
```
