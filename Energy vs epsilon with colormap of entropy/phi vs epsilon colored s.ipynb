```julia
using Plots
using CSV
using DelimitedFiles, Plots
file = raw"plot_data0000.txt"
M = readdlm(file)
phi_f = M[:,1];
eigen = M[:,2];
s = M[:,3];
using Plots
gr()
MyTitle = "L = 6, colormap = Entanglement entropy";
p = plot(phi_f,eigen,
    seriestype = :scatter,
    markerstrokecolor = "grey30",
    markerstrokewidth=0.0168,
    markersize=1.6,
    thickness_scaling = 1.4,
    #xlims=(0,0.20), 
    title = MyTitle,
    label = "",
    dpi=600,
    zcolor = s,
    color = :RdBu_4,
    right_margin = 5Plots.mm,
    left_margin = Plots.mm,
    titlefontsize=10,
    guidefontsize=10,
    tickfontsize=13,
    legendfontsize=10,
    )
plot!(size=(900,700))
hline!([[-2.8909]],lc=:green,legend=false)
hline!([ [0]],lc=:green,legend=false)
hline!([ [2.8909]],lc=:green,legend=false)
xlabel!("epsilon")
ylabel!("Energy")
```
